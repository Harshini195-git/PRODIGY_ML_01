```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn import metrics
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/mlflow/mlflow-example/refs/heads/master/wine-quality.csv')
print(df.head())
```

```python
df.info()
```

```python
df.describe().T
```

```python
df.isnull().sum()
```

```python
for col in df.columns:
  if df[col].isnull().sum() > 0:
    df[col] = df[col].fillna(df[col].mean())

df.isnull().sum().sum()
```

```python
df.hist(bins=20, figsize=(10, 10))
plt.show()
```

```python
plt.bar(df['quality'], df['alcohol'])
plt.xlabel('quality')
plt.ylabel('alcohol')
plt.show()
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        try:
            df[col] = pd.to_numeric(df[col], errors='coerce')
        except:
            pass

plt.figure(figsize=(12, 12))
sb.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.show()
```

```python
df = df.drop('total sulfur dioxide', axis=1)
df['best quality'] = [1 if x > 5 else 0 for x in df.quality]
df.replace({'white': 1, 'red': 0}, inplace=True)
```

```python
features = df.drop(['quality', 'best quality'], axis=1)
features = features.fillna(features.mean())
target = df['best quality']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(features, target, test_size=0.2, random_state=40)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
xtrain = imputer.fit_transform(xtrain)
xtest = imputer.transform(xtest)
xtrain.shape, xtest.shape
```

```python
norm = MinMaxScaler()
xtrain = norm.fit_transform(xtrain)
xtest = norm.transform(xtest)
```

```python
models = [LogisticRegression(),SVC(kernel='rbf')]
```

```python
models = [LogisticRegression(), SVC()]

for model in models:
    model.fit(xtrain, ytrain)
    print("\nModel:", model)
    print("Training AUC :", metrics.roc_auc_score(ytrain, model.predict(xtrain)))
    print("Validation AUC :", metrics.roc_auc_score(ytest, model.predict(xtest)))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm = confusion_matrix(ytest, models[1].predict(xtest))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=models[1].classes_)
disp.plot()
plt.show()
```

```python
print(metrics.classification_report(ytest,models[1].predict(xtest)))
```
